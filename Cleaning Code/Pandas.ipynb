## Treating null values
### size

```python
data["Category"].unique()
```

```python
data.groupby("Category")["Size"].unique()
```

```python
data["Size"].isnull().sum()
```

```python
#treating null values  or replacing all values in the electronic  category 
data.loc[data["Category"]=="Electronics" , "Size"] = "Not applicable"
```

```python
#treating null values  or replacing all values in the Accessories  category 
data.loc[data["Category"]=="Accessories" , "Size"] = "Free Size"
```

```python
#treating null values  or replacing all values in the Footwear  category 
data.loc[data["Category"]=="Footwear" , "Size"] = "Not available"
```

```python
#treating null values  or replacing all values in the clothing category 
data.loc[(data["Category"]=="Clothing") & (data["Size"].isnull()) ,"Size"] = data[data["Category"]=="Clothing"]["Size"].mode()[0]
```

## Review Rating


```python
data.groupby("Item Purchased")["Review Rating"].mean()
```

```python
# filling null values  of Review Rating  by the mean the product 
data["Review Rating"] = data["Review Rating"].fillna(data.groupby("Item Purchased")["Review Rating"].transform("mean")).round(1)
```

## Previous Purchase


```python
# dilling null values by 0 in previous puchase
data["Previous Purchases"] = data["Previous Purchases"].fillna(0)
```

## Purchased Amount (USD)


```python
data.groupby("Item Purchased")["Purchase Amount (USD)"].mean()
```

```python
data["Purchase Amount (USD)"] = data["Purchase Amount (USD)"].fillna(data.groupby("Item Purchased")["Purchase Amount (USD)"].transform("mean"))
```
